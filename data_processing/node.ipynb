```python
def search_type_industry(type, industry):
    if type == 'all' and industry == 'all':
        return dataset
    elif type == 'all':
        return dataset[dataset['industry']==industry]
    elif industry == 'all':
        return dataset[dataset['type']==type]
    else:
        return dataset[(dataset['type']==type) & (dataset['industry']==industry)]


#  input the type and industry for searching
#  type and industry should be string type
#  e.g. type = "Domain"
#  e.g. industry = "['C', 'E', 'A', 'B']"
typeName = "Domain"
industryName = "['C', 'E', 'A', 'B']"
result = search_type_industry(typeName, industryName)
result
```
